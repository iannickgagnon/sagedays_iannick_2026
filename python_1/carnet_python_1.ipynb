# Exercice - Racines de $ax+bx+c$

```python
a = 1
b = 2
c = -1

r1 = (-b - (b**2 - 4 * a * c)**0.5) / (2 * a)
r2 = (-b + (b**2 - 4 * a * c)**0.5) / (2 * a)

print(r1, r2)
```

```python
def calculer_afficher_racines(a, b, c):

  # Calcul des racines
  r1 = (-b - (b**2 - 4 * a * c)**0.5) / (2 * a)
  r2 = (-b + (b**2 - 4 * a * c)**0.5) / (2 * a)

  # Affichage
  print(r1, r2)
```

```python
calculer_afficher_racines(1, 2, 2)
```
