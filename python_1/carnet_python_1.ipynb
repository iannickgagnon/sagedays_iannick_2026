# Slides 10-12 - Conventions d'écriture et commentaires d'en-tête

````python
def calculer_afficher_racines_2(a, b, c):
  """
  Calcule et affiche les racines d'une quadratique.

  Arguments:
    a (int): Coefficient du terme x^2.
    b (int): Coefficient du terme x^1.
    c (int): Coefficient du terme x^0.

  Retourne:
    Rien.
  """

  # Calcul des racines
  racine_discriminant = (b**2 - 4 * a * c)**0.5
  r1 = (-b - racine_discriminant) / (2 * a)
  r2 = (-b + racine_discriminant) / (2 * a)

  # Affichage
  print(r1, r2)
````

# BONUS - Annotations de types et validation


Ici, j'importe l'annotation `Union` de la librairie `typing` pour signaler que les paramètres sont soit des entiers (`int`) ou des rationnels (`float`). J'utilise aussi l'annotation `Final` pour signaler q'il s'agit d'une constante.

* L'annotation s'insère à même l'en-tête suivant un deux-points après les paramètres (`param: type`).

* L'annotation du type de retour se fait avec une flèche (`->`) suivant la parenthèse fermante (`def foo(param: type_param) -> type_retour: ...`).

* `None` représente **l'absence d'une valeur**.

* Permis d'annoter les déclarations de la même manière que l'on annote les paramètres (`variable: type = valeur`).

* Lorsqu'un paramère possède une valeur par défaut, la mention `optionnel` est ajoutée après l'annotation de type.

# Commentaire éditorial:
> Certain(e)s vous diront que c'est **trop verbeux** ou pas strictement idiomatique (c.-à-d. ne suit pas *exactement* les conventions), mais **cela ne ralentit pas le code**. Je favorise *toujours* la clarté sémantique en premier. Je vous encourage à trouver votre voix dans ce débat.

````python
from typing import Union, Final

def calculer_afficher_racines(a: Union[int, float],
                              b: Union[int, float],
                              c: Union[int, float],
                              epsilon: float = 1e-12) -> None:
  """
  Calcule et affiche les racines d'une quadratique.

  Arguments:
    a (Union[int, float]): Coefficient du terme x^2.
    b (Union[int, float]): Coefficient du terme x^1.
    c (Union[int, float]): Coefficient du terme x^0.
    epsilon (float, optionnel): Tolérance numérique. Valeur par défaut de 1e-12.

  Retourne:
    Rien.

  Erreurs:
    ValueError: Lorsque |a| est inférieur à la 'epsilon'.
  """

  # Valider qu'il s'agit d'une quadratique
  if abs(a) < epsilon:
    raise ValueError("Le coefficient 'a' doit être non nul pour une équation quadratique.")

  # Calcul des racines
  racine_discriminant: Final[float] = (b ** 2 - 4 * a * c) ** 0.5
  r1: Final[float] = (-b - racine_discriminant) / (2 * a)
  r2: Final[float] = (-b + racine_discriminant) / (2 * a)

  # Affichage
  print(r1, r2)
````

**REMARQUE:** Les explications qui suivent sont **facultatives** et pour votre **enrichissement personnel**.

* La docstring de la fonction `calculer_quotient_reste` contient un bloc `Erreurs:` qui décrit les erreurs qui sont levées explicitement par celle-ci avec une instruction `raise`.

* La fonction `isinstance(obj, type)` vérifie si `obj` est une *instance* de la *classe* `type`. Dans ce cas-ci, on vérifie si les deux paramètres sont des entiers, donc de la classe `int`.

En mots: Si l'un ou l'autre `dividende` et `diviseur` n'est pas un entier, soulève une erreur de type.



````python
from typing import Final

def calculer_quotient_reste(dividende: int, diviseur: int) -> tuple[int, int]:
  """
  Effectue la division entière entre deux entiers.

  Arguments:
    dividende (int): Le dividende.
    diviseur (int): Le diviseur.

  Retourne:
    Un tuple contenant:
      - int: Le quotient de la division entière.
      - int: Le reste de la division entière.

  Erreurs:
    TypeError: Si l'un ou l'autre des paramètres n'est pas un antier.
  """

  # BONUS: Valider les types des paramètres fournis
  if not isinstance(dividende, int) or not isinstance(diviseur, int):
    raise TypeError(f'Le dividende ({dividende}) et le diviseur ({diviseur}) doivent être des entiers.')

  # Calculer le quotient et le reste
  quotient: Final[int] = dividende // diviseur
  reste: Final[int] = dividende % diviseur

  return quotient, reste


def afficher_quotient_reste():
  """
  Saisit deux entiers et puis affiche le quotient et le reste.

  Arguments:
    Aucuns.

  Retourne:
    Rien.
  """

  # BONUS: Saisir dividende et diviseur
  dividende: Final[int] = int(input("Saisissez le dividende: "))
  diviseur: Final[int] = int(input("Saisissez le diviseur: "))

  # Effectuer calcul
  quotient, reste = calculer_quotient_reste(dividende, diviseur)

  # Affichage
  print('Le quotient vaut', quotient)
  print('Le reste vaut', reste)


# Faire appel à la procédure principale
afficher_quotient_reste()
````

# Slide no.14 - EXERCICE: Minimum et maximum

Implémentez une fonction nommée `obtenir_min_max` qui est appelée à l'intérieur d'une fonction nommée `afficher_min_max` dont le comportement est le suivant :

```python
>>> afficher_min_max(10, 1)
Le minimum vaut 1
Le maximum vaut 10
```

**Aide:** [https://docs.python.org/fr/3.14/library/functions.html](https://docs.python.org/fr/3.14/library/functions.html)

````python
def obtenir_min_max(valeur_1: int, valeur_2: int) -> tuple[int, int]:
  """
  Détermine le minimum et le maximum entre deux entiers.

  Arguments:
    valeur_1 (int): La première valeur fournie.
    valeur_2 (int): La deuxième valeur fournie.

  Retourne:
    Un tuple contenant:
      - int: La plus petite des deux valeurs.
      - int: La plus grande des deux valeurs.
  """

  # BONUS: Valider les types des paramètres fournis
  if not isinstance(valeur_1, int) or not isinstance(valeur_1, int):
    raise TypeError(f'Les valeurs fournies doivent être des entiers.')

  # Déterminer min et max
  valeur_min = min(valeur_1, valeur_2)
  valeur_max = max(valeur_1, valeur_2)

  return valeur_min, valeur_max


def afficher_min_max(valeur_1: int, valeur_2: int) -> None:
  """
  Affiche le minimum et le maximum entre deux entiers.

  Arguments:
    valeur_1 (int): La première valeur fournie.
    valeur_2 (int): La deuxième valeur fournie.

  Retourne:
    Rien.
  """

  # Séparer le minimum et le maximum
  valeur_min, valeur_max = obtenir_min_max(valeur_1, valeur_2)

  # Affichage
  print('Le minimum vaut', valeur_min)
  print('Le maximum vaut', valeur_max)


# Faire appel à la procédure principale
afficher_min_max(10, 1)
````

# Slide no.25 - EXERCICE: Somme (im)paire

````python
from typing import Union

def calculer_somme_paire_impaire(liste_ou_tuple: Union[list[int], tuple[int]]) -> tuple[int, int]:
  """
  Calcule les sommes paires et impaires d'une collection d'entiers.

  Args:
    liste_ou_tuple (Union[list[int], tuple[int]]): Collection d'entiers.

  Retourne:
    Tuple contenant:
      - int: La somme des valeurs paires.
      - int: La somme des valeurs impaires.
  """

  # Initialiser sommes
  somme_paire: int = 0
  somme_impaire: int = 0

  # Calculer sommes
  for valeur in liste_ou_tuple:
    if valeur % 2 == 0:
      somme_paire += valeur
    else:
      somme_impaire += valeur

  return somme_paire, somme_impaire
````

# Slide no.27 - EXERCICE: Modulo fait maison

````python
from math import copysign

def modulo(dividende: int, diviseur: int) -> int:
  """
  Calcule le modulo entre le dividende et le diviseur.

  NOTE: Cette version est plus compliquée, car elle tient compte des signes.

  Args:
    dividende (int): Le dividende.
    diviseur (int): Le diviseur.

  Retourne:
    int: Le modulo.

  Erreurs:
    ValueError: Si le dividende et le diviseur n'ont pas le même signe.
  """

  # BONUS: Valider les entrées
  if dividende * diviseur < 0:
    raise ValueError("Les valeurs doivent être du même signe")

  # Extraire le signe du résultat attendu
  signe: float = copysign(1, dividende)

  # Utiliser les valeurs absolues
  abs_dividende: int = abs(dividende)
  abs_diviseur: int = abs(diviseur)

  # Soustractions successives
  reste = abs_dividende
  while reste > 0:
    reste -= abs_diviseur

  return int(signe * (reste + abs_diviseur))
````

# Slide no.30 - EXERCICE: Moyenne et écart-type

````python
from typing import Union

def calculer_moyenne(valeurs: list[Union[int, float]]) -> float:
  """
  Calcule la moyenne d'une liste de nombres.

  Args:
    valeurs (list[Union[int, float]]): Liste de nombres.

  Retourne:
    float: La moyenne.
  """

  # Initialiser la somme
  somme: float = 0.0

  # Additionner les termes
  for valeur in valeurs:
    somme += valeur

  # Calculer et retourner la moyenne
  return somme / len(valeurs)
````

````python
def calculer_ecart_type(valeurs: list[Union[int, float]]) -> float:
  """
  Calcule l'écart-type d'une liste de nombres.

  Args:
    valeurs (list[Union[int, float]]): Liste de nombres.

  Retourne:
    float: L'écart-type.
  """

  # Calculer la moyenne
  moyenne: float = calculer_moyenne(valeurs)

  # Calculer la somme interne
  somme: float = 0.0
  for valeur in valeurs:
    somme += (valeur - moyenne) ** 2

  # Calculer et retourner l'écart-type
  return (somme / len(valeurs)) ** 0.5
````

````python
def saisir_valeurs() -> list[float]:
  """
  Saisit des valeurs numériques jusqu'à ce que l'utilisateur quitte.

  Args:
    Aucuns.

  Retourne:
    list[float]: Les valeurs saisies par l'utilisateur/trice.
  """

  # Valeur qui met fin à la saisie
  QUITTER: str = ''

  # Initialiser conteneur
  valeurs: list[Union[int, float]] = []

  # Saisir
  while True:
    valeur_str: str = input("Saisissez une valeur (Entrée pour quitter): ")
    # Quitter si vide
    if valeur_str == QUITTER:
      break
    valeurs.append(float(valeur_str))

  return valeurs
````

````python
def afficher_moyenne_ecart_type() -> None:
  """
  Affiche la moyenne et l'écart-type d'une liste de nombres.

  Args:
    Aucuns.

  Retourne:
    Rien.
  """

  # Saisie
  valeurs: list[float] = saisir_valeurs()

  # Calculer la moyenne et l'écart-type
  moyenne: float = calculer_moyenne(valeurs)
  ecart_type: float = calculer_ecart_type(valeurs)

  # Afficher
  print('Moyenne    :', moyenne)
  print('Écart-type :', ecart_type)
````
