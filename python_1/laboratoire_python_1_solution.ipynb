    # Boucle infinie
    while True:

````python
def obtenir_jour_semaine_int(jour: int, mois: int, annee: int) -> int:
    """
    Retourne le jour de la semaine associé à une date sous la forme d'un entier.

    Source: https://en.wikipedia.org/wiki/Determination_of_the_day_of_the_week

    Les jours de la semaine sont codés comme suit:
      0 = dimanche
      1 = lundi
      2 = mardi
      3 = mercredi
      4 = jeudi
      5 = vendredi
      6 = samedi

    Args:
      jour (int): Jour du mois.
      mois (int): Mois de l'année, de 1 à 12.
      annee (int): Année.

    Returns:
      int: Le jour de la semaine sous forme d'un entier de 0 à 6.
    """

    # NOTE: Nous faisons confiance aux valeurs fournies par l'utilisateur/trice

    if mois < 3:
      mois += 12
      annee -= 1

    return ((23 * mois) // 9 + jour + 4 + annee + annee // 4 - annee // 100 + annee // 400 - 2) % 7
````

````python
def obtenir_nom_jour_semaine(jour: int, mois: int, annee: int) -> str:
    """
    Retourne le jour de la semaine associé à une date sous la d'une chaîne de caractères.

    Les jours de la semaine sont codés comme suit:
      0 = dimanche
      1 = lundi
      2 = mardi
      3 = mercredi
      4 = jeudi
      5 = vendredi
      6 = samedi

    Args:
      jour (int): Jour du mois.
      mois (int): Mois de l'année, de 1 à 12.
      annee (int): Année.

    Returns:
      str: Le nom du jour de la semaine.
    """

    # Association clé:valeur de la forme 0 -> 'lundi', etc.
    CONVERSION_JOUR_ENTIER_NOM: dict[int, str] = {
      0: 'lundi',
      1: 'mardi',
      2: 'mercredi',
      3: 'jeudi',
      4: 'vendredi',
      5: 'samedi',
      6: 'dimanche',
    }

    # Obtenir le jour de la semaine sous la forme d'un entier allant de 0 à 6
    jour_semaine_int: int = obtenir_jour_semaine_int(jour=jour, mois=mois, annee=annee)

    # Convertir en nom de semaine et retourner
    return CONVERSION_JOUR_ENTIER_NOM[jour_semaine_int]
````
